Difference between built-in functions and user-defined functions:
Built-in functions: These functions are provided as part of the Python language itself. They are ready to use and do not require any additional code or definitions. Built-in functions are typically used for common tasks and operations in Python. Examples of built-in functions include print(), len(), and max(). Here's an example:
python
Copy code
#Built-in function example
length = len("Hello, world!")
print(length)  # Output: 13
User-defined functions: These functions are created by the users themselves to perform specific tasks or operations. Users define the function by providing a name, input parameters, and the desired functionality. User-defined functions can be customized to suit specific requirements. Here's an example:
python
Copy code
#User-defined function example
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 3)
print(result)  # Output: 8

Passing arguments to a function in Python:
Positional arguments: These arguments are passed to a function based on their position or order. The arguments are matched with the function parameters in the same order. The number and position of arguments must match the function's parameter list.
Example:

python
Copy code
def greet(name, age):
    print("Hello", name, "! You are", age, "years old.")

greet("Alice", 25)
Output:

sql
Copy code
Hello Alice ! You are 25 years old.

Keyword arguments: These arguments are passed to a function using the argument name along with the value. The order of the arguments does not matter in this case, as they are matched based on their names. Keyword arguments provide more flexibility when calling a function.
Example:

python
Copy code
def greet(name, age):
    print("Hello", name, "! You are", age, "years old.")

greet(age=30, name="Bob")
Output:

sql
Copy code
Hello Bob ! You are 30 years old.

Purpose of the return statement in a function:
The return statement in a function is used to specify the value that the function should return when it is called. It allows the function to provide an output or result back to the calling code. The return statement also terminates the execution of the function.

A function can have multiple return statements, but only one of them will be executed. When a return statement is encountered, the function exits immediately, and the specified value is returned.

Example:

python
Copy code
def find_max(numbers):
    if len(numbers) == 0:
        return None

    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num

    return max_num

numbers_list = [5, 2, 8, 3, 10]
max_number = find_max(numbers_list)
print(max_number)  # Output: 10
In this example, the find_max() function returns the maximum number from a list of numbers. The first return None statement handles the case when the input list is empty. The subsequent return max_num statement is executed.

Lambda functions in Python:
Lambda functions, also known as anonymous functions, are small and inline functions that are defined without a name. They are created using the lambda keyword and can take any number of arguments but can only have one expression.
Example:

python
Copy code
#Regular function to square a number
def square(x):
    return x**2

result = square(5)
print(result)  # Output: 25

#Equivalent lambda function to square a number
square_lambda = lambda x: x**2

result_lambda = square_lambda(5)
print(result_lambda)  # Output: 25
Lambda functions are useful when a small, one-line function is required and defining a regular function would be unnecessary.

Scope in Python functions:
Local scope: Variables defined inside a function have local scope. They are only accessible within the function itself. Local variables are created when the function is called and destroyed when the function returns or terminates.

Global scope: Variables defined outside any function or at the top level of a module have global scope. They are accessible from anywhere in the code, including inside functions. Global variables are created when the module is imported or when they are explicitly defined, and they persist throughout the program execution.

Example:

python
Copy code
global_var = 5  # Global variable

def my_function():
    local_var = 10  # Local variable
    print(local_var)  # Output: 10
    print(global_var)  # Output: 5

my_function()
print(global_var)  # Output: 5
In this example, local_var is a local variable within the my_function() function, while global_var is a global variable defined outside any function.

Answer for Q.6. :- 

Using the return statement to return multiple values:
In Python, a function can return multiple values by separating them with commas in the return statement. The values are returned as a tuple, which can then be unpacked or accessed as needed.

Example:

python
Copy code
def get_name_info(name):
    uppercase_name = name.upper()
    lowercase_name = name.lower()
    return uppercase_name, lowercase_name

result = get_name_info("John")
print(result)  # Output: ('JOHN', 'john')

#Unpacking the returned tuple
uppercase, lowercase = get_name_info("John")
print(uppercase)  # Output: 'JOHN'
print(lowercase)  # Output: 'john'
In this example, the get_name_info() function returns the uppercase and lowercase versions of a given name as a tuple. The tuple is then unpacked into separate variables uppercase and lowercase.

Difference between "pass by value" and "pass by reference" concepts:
In Python, all function arguments are passed by value, but the behavior can appear similar to pass by reference in some cases.

Pass by value: When a function is called with a variable as an argument, a copy of the value is created and passed to the function. Any modifications made to the parameter variable inside the function do not affect the original variable outside the function.

Pass by reference (continued):

In Python, there is no explicit pass by reference. However, when passing mutable objects like lists or dictionaries as arguments, the reference to the object is passed. This means that changes made to the object inside the function will be reflected outside the function as well.

Example:

python
Copy code
def modify_list(lst):
    lst.append(4)  # Modifying the list

my_list = [1, 2, 3]
modify_list(my_list)
print(my_list)  # Output: [1, 2, 3, 4]
In this example, the modify_list() function modifies the list by appending the value 4 to it. This change is visible outside the function, indicating that the list was passed by reference.

Code to create a function that takes a full name as an argument and returns the first name and last name:
python
Copy code
def extract_name(full_name):
    name_parts = full_name.split()
    first_name = name_parts[0]
    last_name = name_parts[-1]
    return first_name, last_name

name = "John Doe"
first, last = extract_name(name)
print("First name:", first)
print("Last name:", last)
In this code, the extract_name() function takes a full name as an argument. It splits the name using the split() method to separate the first name and last name. The first name is accessed using index 0, and the last name is accessed using index -1 (to handle multiple names). The function returns the first name and last name as a tuple, which is then unpacked into separate variables and printed.
